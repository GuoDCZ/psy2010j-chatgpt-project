```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter

Questions = (
    "All_s1",
    "All_s2",
    "None_s1",
    "None_s2",
    "Some_s",
    "All_d",
    "None_d",
    "Some_d"
)
Counts = {
#     "Correct": np.array([98, 100, 100, 96, 97, 99, 95, 90]),
    "Unclear": np.array([2, 0, 0, 4, 3, 1, 0, 0]),
    "Wrong": np.array([0, 0, 0, 0, 0, 0, 5, 10])
}

fig, ax = plt.subplots(figsize=(7, 2))
ax.set_ylim(0, 12)
bottom = np.zeros(8)

c = ['b','r']
i = 0
for boolean, count in Counts.items():
    p = ax.bar(Questions, count, 0.7, label=boolean, bottom=bottom, color=c[i], alpha=0.3)
    bottom += count
    i += 1
    
ax.set_title("Answer from ChatGPT to questions on basic quantifiers")
ax.legend(loc="upper left")


# Function to format y-axis tick labels as percentages
def percentage_formatter(x, pos):
    return f"{x:.1f}%"

# Set the custom formatter for y-axis tick labels
ax.yaxis.set_major_formatter(FuncFormatter(percentage_formatter))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2.-0.2, height+5, '%s%%'%int(height), size=10)

Questions = (
    "all",
    "all_not",
    "some",
    "some_not",
    "none",
    "none_not",
)
Counts = {
    "Wrong Answer": np.array([24, 8, 2, 48, 0, 100])
}

fig, ax = plt.subplots(figsize=(5, 2))
ax.set_ylim(0, 120)
bottom = np.zeros(6)

c = ['r']
i = 0
for boolean, count in Counts.items():
    p = ax.bar(Questions, count, 0.7, label=boolean, bottom=bottom, color=c[i], alpha=0.3)
    autolabel(p)
    bottom += count
    i += 1
    
ax.set_title("Answers from ChatGPT to questions on negation")
ax.legend(loc="upper left")


# Function to format y-axis tick labels as percentages
def percentage_formatter(x, pos):
    return f"{x:.0f}%"

# Set the custom formatter for y-axis tick labels
ax.yaxis.set_major_formatter(FuncFormatter(percentage_formatter))

plt.show()
```
